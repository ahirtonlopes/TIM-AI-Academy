Objetivo: Vamos criar gráficos para explorar o comportamento de clientes, como a distribuição de chamadas e mensagens de texto, e visualizar tendências.

1. Distribuição do Número de Chamadas por Dia: Um histograma que mostra a quantidade de chamadas realizadas pelos clientes ao longo de um mês.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Exemplo de dataset de telecom
data = {'dia': np.arange(1, 31),
        'chamadas': np.random.randint(50, 200, size=30)}

df = pd.DataFrame(data)

# Criação do histograma de chamadas por dia
plt.figure(figsize=(10, 6))
plt.bar(df['dia'], df['chamadas'], color='royalblue')
plt.xlabel('Dia do Mês')
plt.ylabel('Número de Chamadas')
plt.title('Distribuição de Chamadas por Dia (Empresa de Telecom)')
plt.show()
```

2. Evolução do Uso de Dados Móveis: Um gráfico de linhas que mostra a quantidade de dados móveis consumidos por usuários durante o mês.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Exemplo de dataset de telecom
data = {'dia': np.arange(1, 31),
        'chamadas': np.random.randint(50, 200, size=30)}

df = pd.DataFrame(data)

# Adicionando a coluna de consumo de dados móveis
df['dados_moveis_gb'] = np.random.uniform(1, 5, size=30)

# Criação do gráfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(df['dia'], df['dados_moveis_gb'], marker='o', color='green')
plt.xlabel('Dia do Mês')
plt.ylabel('Dados Móveis Consumidos (GB)')
plt.title('Evolução do Consumo de Dados Móveis (Empresa de Telecom)')
plt.grid(True)
plt.show()
```

3. Relação entre Chamadas e Mensagens de Texto: Um gráfico de dispersão que compara o número de chamadas com o número de mensagens de texto enviadas pelos clientes.

```python
# Exemplo de dataset de telecom com coluna 'mensagens'
data = {'dia': np.arange(1, 31),
        'chamadas': np.random.randint(50, 200, size=30),
        'dados_moveis_gb': np.random.uniform(1, 5, size=30)}

df = pd.DataFrame(data)

# Adicionando a coluna 'mensagens' corretamente
df['mensagens'] = np.random.randint(100, 500, size=30)

# Cenário 3: Comparação entre chamadas, dados móveis e mensagens
plt.figure(figsize=(10, 6))
plt.plot(df['dia'], df['chamadas'], label='Chamadas', marker='o')
plt.plot(df['dia'], df['dados_moveis_gb'], label='Dados Móveis (GB)', marker='o')
plt.plot(df['dia'], df['mensagens'], label='Mensagens', marker='o')
plt.xlabel('Dia do Mês')
plt.ylabel('Quantidade')
plt.title('Comparação de Chamadas, Dados Móveis e Mensagens (Empresa de Telecom)')
plt.legend()
plt.grid(True)
plt.show()
```
