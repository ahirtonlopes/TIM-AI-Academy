```python
# Carregar o dataset (por hora, exemplo fictício)
df = pd.read_csv('telecom_data.csv')
```

```python
# Exibir as primeiras linhas do dataframe
print(df.head())
```

```python
# Resumo estatístico das variáveis numéricas
print(df.describe())
```

```python
# Verificar valores faltantes
print(df.isnull().sum())
```

Distribuição de Idade

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['idade'], bins=30, kde=True)
plt.title('Distribuição da Idade dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Churn por Gênero

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='genero', hue='churn')
plt.title('Churn por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.legend(title='Churn', loc='upper right', labels=['Não', 'Sim'])
plt.show()
```

Consumo Mensal de Dados

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='churn', y='consumo_dados_gb', data=df)
plt.title('Consumo Mensal de Dados por Status de Churn')
plt.xlabel('Churn')
plt.ylabel('Consumo de Dados (GB)')
plt.xticks([0, 1], ['Não', 'Sim'])
plt.show()
```

Tempo de Permanência

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['tempo_permanencia_meses'], bins=30, kde=True)
plt.title('Distribuição do Tempo de Permanência dos Clientes')
plt.xlabel('Tempo de Permanência (Meses)')
plt.ylabel('Frequência')
plt.show()
```
